# 5.) 
Consider the following array:\
sampleArray \= numpy\.array(\[\[34,43,73\],\[82,22,12\],\[53,94,66\]\])\
i. Sort above array by second row\
ii. Sort above array by second column\
iii. Print max from axis 0 and min from axis 1\
iv. Delete col 2 and insert new Column numpy.array(\[\[10,10,10]]) in its place

```

import numpy as np

sampleArray = np.array([[34, 43, 73], [82, 22, 12], [53, 94, 66]])
print(f"Sample Array ::\n{sampleArray}")

# i. Sort above array by second row

print(f"After sorting above array by second row ::\n{sampleArray[:, sampleArray[1].argsort()]}")

# ii. Sort above array by second column

print(f"After sorting above array by second column ::\n{sampleArray[sampleArray[:, 1].argsort(), :]}")

# iii. Print max from axis 0 and min from axis 1

print(f"Max from axis 0 : {np.max(sampleArray, axis=0)}\nMin from axis 1 : {np.min(sampleArray, axis=1)}")

# iv. Delete col 2 and insert new Column numpy.array([[10,10,10]]) in its place

sampleArray = np.delete(sampleArray, 1, axis=1)
print(f"After deleting col 2 from above array ::\n{sampleArray}")
sampleArray = np.insert(sampleArray, 1, np.array([[10, 10, 10]]), 1)
print(f"After inserting np.array([[10,10,10]]) in col 2 ::\n{sampleArray}")
```

# 6.) 
Remove all the elemnets from an array that exist in another array?


```
import numpy as np
x1 = np.random.randint(10, size=6)
x2 = np.random.randint(10, size=6)
print(f"First Array : {x1}")
print(f"Second Array : {x2}")
x2 = np.unique(x2)
for i in x2:
    if i in x1:
        x1 = np.delete(x1, np.where(x1 == i))
print(f"First Array after removing elements from Second Array : {x1}")
```

# 7.)
Swap two columns in a 2d NumPy array.


```
import numpy as np
a = np.array(np.arange(1, 10)).reshape(3, 3)
print(f"2D-Array ::\n{a}")
temp = a[:,0].copy(); a[:,0]=a[:, 2];a[:,2]=temp
print(f"2D-Array after swapping column 0 and column 2 ::\n{a}")
```

# 8.)
Swap two rows in a 2d NumPy array.


```
import numpy as np
a = np.array(np.arange(1, 10)).reshape(3, 3)
print(f"2D-Array ::\n{a}")
temp = a[0, :].copy(); a[0, :]=a[2, :];a[2, :]=temp
print(f"2D-Array after swapping row 0 and row 2 ::\n{a}")
```

# 14.) 
For a 1D array with numeric values, find minimum, maximum, mean, median, standard deviation, 5th and 95th percentile, unique values, count of unique values, and the most frequent value.


```
import numpy as np
n = int(input("Enter the number of elements in array : "))
arr = np.zeros((n), dtype='int')
for i in range(n):
    arr[i] = int(input("Enter a value : "))

print(f"Minimum value : {np.amin(arr)}")
print(f"Maximum value : {np.amax(arr)}")
print(f"Mean value : {np.mean(arr)}")
print(f"Median value : {np.median(arr)}")
print(f"Standard Deviation value : {np.std(arr)}")
print(f"5th Percentile value : {np.percentile(arr, 5)}")
print(f"95th Percentile value : {np.percentile(arr, 95)}")
u = np.unique(arr)
print(f"Unique values : {u}")
print(f"Unique values count : {u.size}")
mx_count = 0
mx_count_value = 0
for i in u:
    i_count = np.count_nonzero(arr == i)
    if mx_count < i_count:
        mx_count = i_count
        mx_count_value = i
print(f"Most Frequent value : {mx_count_value}")
```

# 15.)
Write a NumPy program to create a 10x10 matrix, in which all the elements on the borders should be equal to 1, and rest others should be 0.


```
import numpy as np
mat = np.zeros((10, 10), dtype='int')
mat[0, :] = 1
mat[:, 0] = 1
mat[9, :] = 1
mat[:, 9] = 1
print(f"The required 10x10 matrix ::\n{mat}")
```

# 17.) 
Count the number of elements in a numpy array which are greater than 10. Further, multiply all such elements with value 10. 


```
import numpy as np
import numpy as np
a = np.random.randint(20, size=10)
print(f"Array ::\n{a}")

print(f"The number of elements which are greater than 10 : {a[np.nonzero(a>10)].size}")

a[np.nonzero(a>10)] *= 10

print(f"Array after multiplying all the elements with value 10 which are greater than 10 ::\n{a}")
```
