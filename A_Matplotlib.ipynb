```
import matplotlib.pyplot as plt

programming_language = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.barh(programming_language, popularity, height=0.75, edgecolor='b', color=['r', 'y', 'c', 'm', 'g', 'w'])

for index, value in enumerate(popularity):
    plt.text(value+0.15, index -0.1, str(value), fontweight='bold', va='baseline',ha='left')

plt.ylabel("Programming Languages")
plt.xlabel("Popularity")
plt.title("Popularity of Programming Languages")

plt.xlim(0, 25)
plt.show()
```

```
import matplotlib.pyplot as plt

barWidth = 0.25


men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]


r1 = list(range(len(men)))
r2 = [x + barWidth for x in r1]


plt.bar(r1, men, color='g', width=barWidth, label='Men')
plt.bar(r2, women, color='r', width=barWidth, label='Women')


plt.xlabel('Person')
plt.xticks([r + 0.125 for r in range(len(men))], ['G1', 'G2', 'G3', 'G4', 'G5'])
plt.ylabel('Scores')


plt.legend()
plt.title("Scores by person")

plt.show()
```

# 8.)

```
import matplotlib.pyplot as plt

people = ('G1', 'G2', 'G3', 'G4', 'G5', 'G6', 'G7', 'G8')
segments = 4


data = [
    [3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039, 7.1123587, 12.77792868, 3.44773477],
    [11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195, 6.79685302, 7.24578743, 3.69371847],
    [3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 4.63832778, 11.16849999, 8.56883433],
    [4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122]
]
percentages = [[round((data[j][i] / sum(data[j])) * 100) for j in range(segments)] for i in range(len(people))]

y_pos = list(range(len(people)))

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)

colors = 'rgwm'
patch_handles = []
left = [0 for i in range(len(people))]

def left_d(left, d):
    for i in range(len(left)):
        left[i] += d[i]
    return left
for i, d in enumerate(data):
    patch_handles.append(ax.barh(y_pos, d,
                                 color=colors[i % len(colors)], align='center',
                                 left=left))

    left = left_d(left, d)


for j in range(len(patch_handles)):
    for i, patch in enumerate(patch_handles[j].get_children()):
        bl = patch.get_xy()
        x = 0.5 * patch.get_width() + bl[0]
        y = 0.5 * patch.get_height() + bl[1]
        ax.text(x, y, "%d%%" % (percentages[i][j]), ha='center')

ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.set_xlabel('Scores')

plt.show()
```

```
import matplotlib.pyplot as plt

programming_language = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.title("Popularity of Programming Languages", bbox={'facecolor':'0.8', 'pad':10})

wedges, texts, autotexts = plt.pie(popularity, autopct='%1.2F%%', colors=['r', 'y', 'c', 'm', 'g', 'b'])
plt.legend(wedges, programming_language,
          title="Programming Languages",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()

```
