## Dot Product ::

$$ a \cdot b = a^T b = \sum_{d=1}^D a_d b_d $$

Notice that the convention $a^T b$ implies that the vectors are by default column vectors, so the result is 1 by 1, which is a scalar.

````
a = np.array([1, 2])
b = np.array([3, 4])
````

````
# Dot product of 1D-Array
dot = 0
for i, j in zip(a,b):
    dot += i*j
dot
````

````
# Doing same using indices
dot = 0
for i in range(len(a)):
    dot += a[i]*b[i]
dot
````

````
a * b # It will give element wise product
````

````
# We can use this result since the dot product is just the sum of these elements
np.sum(a * b)
````

## Speed Test ::

````
## speed comparison ##
import numpy as np
from datetime import datetime

# note: you can also use %timeit

a = np.random.randn(100)
b = np.random.randn(100)
T = 100000

def slow_dot_product(a, b):
  result = 0
  for e, f in zip(a, b):
    result += e*f
  return result

t0 = datetime.now()
for t in range(T):
  slow_dot_product(a, b)
dt1 = datetime.now() - t0

t0 = datetime.now()
for t in range(T):
  a.dot(b)
dt2 = datetime.now() - t0

print("dt1 / dt2:", dt1.total_seconds() / dt2.total_seconds())
````

Basically the first return value is an array containing all the eigen values and the second return value is an array containing all the eigen vectors organised into a matrix. Ofcourse we can use what we know about eigen decomposition to check that these answers are correct.

````
Lam, v = np.linalg.eig(A)
# Now by convention eigen vectors are stored as column vectors inside v, luckily we just looked at how to select columns of a matrix.
````

````
# So let's check whether the 0th eigen value multiplied by the 0th eigen vector is equal to A times the same eigen vector.
v[:,0] * Lam[0] == A @ v[:,0]
````

So interestingly we get back an array containing one True and one False.
Like many Numpy operators this too operates element wise, both on the left-hand side and right-hand side, return a vector of size 2. And therefore, we are compairing two vectors of size 2 with ==.


Now there's clearly a problem here, since by the rules of eigen decomposition we would expect to see True and True.


Now let's check the values :

````
v[:,0] * Lam[0], A @ v[:,0]
````

Interestingly, the values look like they are the same and yet Numpy thinks they are not equal, ofcourse this is due to numerical precision.
Since computers don't have an infinite amount of memory, they cant have an infinite amount of precision and therefore, numbers have to be rounded-off at some point. That's why you see terms like 32-bit or 64-bit, which tells you how much space each floating point number take off.

So in Numpy, the correct way to compare whether two arrays are equal is to use the Numpy allclose().

````
np.allclose(v[:,0] * Lam[0], A @ v[:,0]) # As expected, it returns True
````

````
# And ofcourse, we could have checked all the eigen vectors at the same time using matrix notation
np.allclose(v @ np.diag(Lam), A @ v)
````

If you know that your matrix is symmetric, then you can use the function : np.linalg.eigh(), which is better for that scenario. The h here stands for hermitian, which is the complex analog of the matrix transpose. It does both a transpose and takes the complex conjugate of the elements.

Symmetric means, $$A = A^T$$


Hermitian means, $$A = A^H$$
$A^H$ = Conjugate transpose of A

In practice, usually you use eigen decomposition on a symmetric matrix like the covariance, so when that's the case you should use eigh instead which has the same API.

## Solving Linear Systems ::

In this, we are going to discuss a very common problem in all areas of science and engineering, which is solving linear systems.


---



Example Problem : The admission fee at a small fiar is \$1.50 for children and \\$4.00 for adults. On a certain day, 2200 people enter the fair, and \\$5050 is collected. How many children and how many adults attended?

So in order to solve this problem, we are going to create two variables, $x_{1}$ and $x_{2}$, to represent the number of children and number of adults respectively.



$$x_{1} + x_{2} = 2200$$
$$1.5x_{1} + 4x_{2} = 5050$$

Since we have a system of 2 equations and 2 unknowns, we can solve this problem.



---


**Linear System in Matrix Form**\
Ofcourse if we let, $x$ be a vector containing $x_{1}$ and $x_{2}$, this becomes a matrix equality. We call this kind of problem a linear system. Luckily we have already covered all the tools we need in order to solve such a problem.


$x =
 \begin{pmatrix}
  x_{1} \\  x_{2}
 \end{pmatrix}$ , $A = \begin{pmatrix}
 1 & 1 \\ 1.5 & 4
 \end{pmatrix}$ , $b =
 \begin{pmatrix}
  2200 \\  5050
 \end{pmatrix}$

 $Ax = b$

As you know the solution to this is to just multiply both sides by $A^{-1}$.
We get, $$ x = A^{-1}b$$


Unfortunately, this is not how you should solve this kind of problem in code. As you recall, the inverse uses an algorithm to find its solution, it turns out that, this algorithm, is both slower and less accurate, than it needs to be.

Linear systems can be solved using a different algorithm that is both faster and more accurate. That algorithm is implemented in np.linalg.solve() function. Thus, whenever you have an expression of the form $A^{-1}b$, you should not literally type this into Numpy, instead you should always use np.linalg.solve().

````
A = np.array([[1, 1], [1.5, 4]])
b = np.array([2200, 5050])
````

````
np.linalg.solve(A, b)
````

````
# Soving in bad way, which is using Matrix Inverse method
np.linalg.inv(A).dot(b)
````

It works, but keep in mind however that this is a very small and simple problem. Sometimes your problem might be 1000's of dimensions and possibly more, in which case you should definitely not use this method.

## The apply() Function ::

The typical use case for apply() function would be similar to the scenario where we want to do some operation on each element of a list.
For example, if we want to square each item, ofcourse in Python we know that for loops are slow, so we will like to avoid them if possible. The apply() function can be used if we want to do the same operation on each row of the dataframe or each column of the dataframe. In other words, it does what you want to do with the for loop wthout actually writing the for loop.\
Let's do an example, suppose i want to have a column called year, where i take the existing date column, parse out the year and convert it into an integer.\
So let's start by writing a function that accepts as input a single row of a dataframe.

````
def date_to_year(row):
    return int(row['date'].split('-')[0])

df.apply(date_to_year, axis=1)
````

* What if we want to assign a new column value where each cell is derived from the values already in its row?\
* Ex. Model interaction between $X1 and X2 ->X1*X2$\
* We can use the apply function!
```
df['x1x2'] = df.apply(lambda row : row['x1']*row['x2'], axis=1)
```
* Pass in axis=1 so the function gets applied across each row instead of each column.\
* Think of it like Python's map function.\
* If you're not familiar with lambda, this is equivalent:

```
def get_interaction(row):
    return row['x1']*row['x2']

df['x1x2'] = df.apply(get_interaction, axis=1)
```
* Function you pass in takes 1 argument, the row.
* Equivalent to:
```
interactions = []
for idx, row in df.iterrows():
    x1x2 = row['x1'] * row['x2']
    interactions.append(x1x2)
df['x1x2'] = interactions
```
