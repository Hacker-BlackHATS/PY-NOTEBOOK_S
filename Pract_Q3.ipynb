```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
read_data = pd.read_csv('wine.csv')
```

```
read_data.hist(column=read_data.columns[1:], figsize=(12,12));

plt.tight_layout();
```

```
for i in range(1, read_data.shape[1]):
    print(f"{read_data.columns[i]} ::\nMean : {read_data[read_data.columns[i]].mean()}\nStandard Deviation : {read_data[read_data.columns[i]].std()}\n")
```

```
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
std_scaler

read_data.iloc[:, 1:] = pd.DataFrame(std_scaler.fit_transform(read_data.iloc[:, 1:]), columns=read_data.columns[1:])
read_data
```

```
for i in range(1, read_data.shape[1]):
    print(f"{read_data.columns[i]} ::\nMean : {int(np.round(read_data[read_data.columns[i]].mean()))}\nStandard Deviation : {int(np.round(read_data[read_data.columns[i]].std()))}\n")
```

```
read_data.hist(column=read_data.columns[1:], figsize=(12,12));

plt.tight_layout();
```
