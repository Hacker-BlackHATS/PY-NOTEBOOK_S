1. Sum of digits of a number

```
def sum_of_digits(number):
    total = 0
    while number > 0:
        total += number % 10
        next_number = number // 10
        number = next_number
    print("Sum of digits :", total)


def input_function():
    number = int(input("Enter a number : "))
    sum_of_digits(number)


if __name__ == "__main__":
    input_function()
```

2. Print the following patterns:

A. 

        1   2   3   4   5
        1   2   3   4 
        1   2   3  
        1   2   
        1  


```
def pattern_a(n):
    for i in range(n + 1, 0, -1):
        for j in range(1, i, +1):
            print(j, end='   ')
        print()


def input_function():
    rows = int(input("Enter number of rows : "))
    pattern_a(rows)


if __name__ == "__main__":
    input_function()
```

B. 

       5
       5   4
       5   4   3
       5   4   3   2
       5   4   3   2   1


```
def pattern_b(n):
    for i in range(n, 0, -1):
        for j in range(n, i-1, -1):
            print(j, end = '   ')     
        print()


def input_function():
    rows = int(input("Enter number of rows : "))
    pattern_b(rows)


if __name__ == "__main__":
    input_function()
```

C. 

		   1   
        2  1   2   
    3   2  1   2   3

```
def pattern_c(n):
    for i in range(n):
        for j in range(n, i+1, -1):
            print(" ", end=' ')
        for k in range(i+1, 0, -1):
            print(k, end=' ')
        for m in range(1, 1+i):
            print(m+1, end=' ')
        print()


def input_function():
    rows = int(input("Enter number of rows : "))
    pattern_c(rows)


if __name__ == "__main__":
    input_function()
```
