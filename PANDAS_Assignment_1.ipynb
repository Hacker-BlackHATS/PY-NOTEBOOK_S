# 2.)
Write a Pandas program to add a column named "column1" in the sixth position of the "coalpublic2013.xlsx" excel sheet and fill it with NaN values.

# 3.)
Write a Pandas program to import given excel data (coalpublic2013.xlsx ) into a dataframe and find details where "Mine Name" starts with "P"

```python
import pandas as pd
import numpy as np

read_file = pd.read_excel('coalpublic2013.xlsx')

select_data = read_file[read_file.Mine_Name.map(lambda x: x.startswith('P'))]

print(select_data)
```

# 4.)
Write a Pandas program to import given excel data (employee.xlsx ) into a Pandas dataframe and find a list of employees where hire_date> 01-01-07.

```python
import pandas as pd
import numpy as np

read_data = pd.read_csv('employee.csv')
read_data.hire_date = pd.to_datetime(read_data.hire_date)

data_coll = read_data[read_data.hire_date > '01-01-07']
print(data_coll)
```

# 5.)
Find the average mileage of each car making company

```python
import pandas as pd
import numpy as np

read_data = pd.read_csv('Automobile_data.csv')

avg_car_mileage = read_data.groupby('company')[['average-mileage']].mean()
print(avg_car_mileage)
```

# 6.)
Find the most expensive car company name

```python
import pandas as pd
import numpy as np

read_data = pd.read_csv('Automobile_data.csv')

expensive_car = read_data[['company', 'price']][read_data.price == np.max(read_data.price)]
print(expensive_car)
```
