**Silhouette Score**

```python
scores = []
for k in k_range:
  km = KMeans(n_clusters=k)
  pre = km.fit_predict(Xscaled)
  s = silhouette_score(Xscaled,pre)
  scores.append(s)
plt.plot(k_range,scores)
plt.xlabel("K values")
plt.ylabel("Sum of squared error");
```
