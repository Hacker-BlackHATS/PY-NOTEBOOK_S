```
import matplotlib.pyplot as plt
```

```
import numpy as np
```

```
x = np.linspace(0, 20, 3600000)
```

```
y = np.sin(x) + 0.2 * x
```

```
plt.plot(x, y)
```

```
plt.plot(3, 2)
```

```
plt.plot(3, 2, 'rs', markersize=20)
```

```
plt.plot(6, 'r<')
```

```
plt.plot([1, 3],[2, 4], 'ro')
```

```
x = [2,4,6,8,10]
y = [3,5,7,9,11]
plt.plot(x, y, 'm^')
```

```
plt.plot(x, y, 'm^-')
```

```
plt.plot(x, y, 'm^:')
```

```
plt.plot(x, y, 'm^:', linewidth=5, markersize=20, markeredgecolor='#aaff00', markerfacecolor = 'r', markeredgewidth=3)
```

```
plt.plot(x, y, 'r*--')
plt.axis([1, 30, 1, 20]) # [xmin, xmax, ymin, ymax]
plt.show()
```

```
plt.plot(x, x, 'r*--', label='X vs X')
y2 = [i**2 for i in x]
plt.plot(x, y2, 'g*--', label='X vs X**2')
plt.legend()
plt.axis([1, 30, 1, 20]) # [xmin, xmax, ymin, ymax]
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('X Vs Y')
plt.grid()
```
