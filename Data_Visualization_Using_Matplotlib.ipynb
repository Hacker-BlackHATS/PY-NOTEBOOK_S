
**Marker**

---


>Marker | Description
>--- | ---
>'o' | Circle
>'*' | Star
>'.' | Point
>',' | Pixel
>'x' | x
>'X' | x (filled)
>'+' | Plus
>'P' | Plus (filled)
>'S' | Square
>'D' | Diamond
>'d' | Diamond (thin)
>'p' | Pentagon
>'H' | Hexagon
>'v' | Triangle Down
>'^' | Triangle Up
>'<' | Triangle Left
>'>' | Triangle Right
>'1' | Tri Down
>'2' | Tri Up
>'3' | Tri Left
>'4' | Tri Right
>'\|' | Vline


---




```
plt.plot(y, marker='X')
plt.show()
```

```
plt.plot(y, marker='X', ms=10) # ms is a short form for markersize
plt.show()
```

```
plt.plot(y, marker='X', ms=10, mec='r') # mec is a short form for markeredgecolor
plt.show()
```

```
plt.plot(y, marker='X', ms=10, mec='r', mfc='y') # mfc is a short form for markerfacecolor
plt.show()
```

### Grid

```
y = [1, 4, 9, 16, 25, 36, 49, 64]
x1 = [1, 16, 30, 42, 55, 68, 77, 88]
x2 = [1, 6, 12, 18, 28, 40, 52, 65]

plt.plot(x1, y, 'ys-')
plt.plot(x2, y, 'go--')
plt.legend(labels=('tv', 'smartphone'), loc='lower right')
plt.grid(True)
plt.title("Ads effects on sales")
plt.xlabel('medium')
plt.ylabel('sales')
plt.show()
```

```
plt.plot(x1, y, 'ys-')
plt.plot(x2, y, 'go--')
plt.legend(labels=('tv', 'smartphone'), loc='lower right')
plt.grid(color='b', ls='-', lw=0.25) # ls stands for linestyle, lw stands for linewidth
plt.title("Ads effects on sales")
plt.xlabel('medium')
plt.ylabel('sales')
plt.show()
```

```
menMeans = [20, 35, 30, 35, 27]
womenMeans = [25, 12, 34, 20, 25]
group = np.arange(5)

plt.bar(group, menMeans, width=0.35, color='r')
plt.bar(group, womenMeans, width=0.35, bottom=menMeans, color='b') # bottom means you can represent the two variables bottom-up
plt.xlabel("Scores")
plt.ylabel("Groups")
plt.title("Scores by group and gender")
plt.xticks(group, ('G1', 'G2', 'G3', 'G4', 'G5')) # Specifies the lables of x-coordinates
plt.yticks(np.arange(0, 81, 10)) # Specifies the lables of x-coordinates
plt.legend(labels=("Men", "Women"))
plt.show() # This graph is showing scores by group and gender
```

```
menMeans = [20, 35, 30, 35, 27]
womenMeans = [25, 12, 34, 20, 25]
group = np.arange(5)

plt.barh(group, menMeans, height=0.35, color='r')
plt.barh(group, womenMeans, height=0.35, left=menMeans, color='b')
plt.ylabel("Scores")
plt.xlabel("Groups")
plt.title("Scores by group and gender")
plt.yticks(group, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.xticks(np.arange(0, 81, 10))
plt.legend(labels=("Men", "Women"))
plt.show() # This graph is showing scores by group and gender
```

### Box

It is important because it can identify mean, median, quantiles and also outliers.

```
np.random.seed(10)
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize = (10, 7)) # This is gone be controlling the size of our graph
plt.boxplot(data)
plt.show()
# These dots are identifying the outliers
```

```
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(80, 30, 200)
data4 = np.random.normal(70, 40, 200)

data = [data1, data2, data3, data4]

fig = plt.figure(figsize = (10, 7))

ax = fig.add_axes([0, 0, 1, 1]) # Controlling the coordinates

bp = ax.boxplot(data)

plt.show()
```

### 3D

```
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0, 1, 100)
x = z * np.sin(20  * z)
y = z * np.cos(20  * z)

ax.plot3D(x, y, z, 'gray')
plt.title('3D line plot')
plt.show()
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(20  * z)
y = z * np.cos(20  * z)
c = x + y


ax.scatter(x, y, z, c=c)
plt.title('3D scatter plot')
plt.show()
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')

x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T
z = np.cos(x ** 2 + y ** 2)

ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
plt.title('Surface plot')
plt.show()
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')

x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T
z = np.cos(x ** 2 + y ** 2)

ax.plot_surface(x, y, z, cmap='nipy_spectral', edgecolor='none')
plt.title('Surface plot')
plt.show()
```
